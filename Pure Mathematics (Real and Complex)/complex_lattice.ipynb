```python
"""complex_lattice.ipynb"""
# Cell 1: Create scatter plot of a complex function

# Imports
from __future__ import annotations

import typing

import matplotlib.pyplot as plt
import numpy as np

if typing.TYPE_CHECKING:
    from matplotlib.axes import Axes
    from numpy.typing import NDArray

%matplotlib widget


def f(z: NDArray[np.complex_]) -> NDArray[np.complex_]:
    """Contains the complex function f(z) = z^2 + iz + 1"""
    return z ** 2 + complex(0, 1) * z + 1


def plot(ax: Axes) -> None:
    """Plot the complex function over given domain"""

    # Create array of real components in interval [-4, 4]
    r_domain: NDArray[np.float_] = np.linspace(-4, 4, 100)
    # Create array of imaginary components (stored as floats) in interval [0, 2]
    j_domain: NDArray[np.float_] = np.linspace(0, 2, 100)
    # Create empty array to hold complex domain values
    z_domain: NDArray[np.complex_] = np.asarray([])
    # Iterate through real vals in r_domain and imaginary vals in j_domain to create function domain
    # Function domain contains all possible combinations of r and j values from the arrays defined above
    for r in r_domain:
        for j in j_domain:
            z_domain = np.append(z_domain, complex(r, j)) 

    # Create range containing f(z) evaluated at each z value in z_domain
    range: NDArray[np.complex_] = f(z_domain)

    # Create scatter plot of real and imaginary components of f(z)
    ax.scatter(np.real(range), np.imag(range))

    # Set aspect ratio to be equal
    ax.set_aspect("equal")

    # Set title and labels
    ax.set_title(r"$f(z) = z^2 + iz + 1, |Re(z)| \leq 4, 0 \leq Im(z) \leq 2$")
    ax.set_xlabel("Real Axis")
    ax.set_ylabel("Imaginary Axis")
    
    # Add grid and axis lines
    ax.grid()
    ax.axhline(0, color="black")
    ax.axvline(0, color="black")
    
        

def main() -> None:
    plt.close("all")
    plt.figure(" ")
    plot(plt.axes())
    plt.show()


main()   
```

```python
# Cell 2: Calculate number of lattice points inside curve for Re(z), Im(z), Re(f(z)), and Im(f(z)) in interval [-10, 10]


# Create new arrays containing Re(z) and Im(z), limited to integers
r_intdomain: NDArray[np.int_] = np.arange(-10, 11)
j_intdomain: NDArray[np.int_] = np.arange(-10, 11)


# Create counter for number of lattice points
lattice_points: int = 0


# Iterate through each integer combination to check whether given conditions are met
for r in r_intdomain:
    for j in j_intdomain:
        # Use algebraic formulas for components of f(z) to check if Re(f(z)) and Im(f(z)) are in interval [-10, 10]
        if abs(r**2 - j**2 - j + 1) <= 10 and abs(2 * r * j + r) <= 10:
            lattice_points += 1


# Print number of lattice points
display(
    f"There are {lattice_points} lattice points inside the curve for the given domain and range restrictions."
)
```
